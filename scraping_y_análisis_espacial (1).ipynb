```python
import requests
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
```

```python
overpass_url =  "http://overpass-api.de/api/interpreter"
query_cafes = """
[out: json];
area["name"="Lima"]->.searchArea;
(
  node["amenity"="cafe"]["name"~"Starbucks", i](area.searchArea);
);
out center;
"""
#Hacer la petición a la API
response = requests.get(overpass_url, params={"data": query_cafes})
data_starbucks = response.json()
```

```python
#EXTRAER INFORMACION Y CONVERTIRLA EN UN DATAFRAME
elements = data_starbucks['elements']
cafes = []

for element in elements:
    if "lat" in element and "lon" in element:
        tags = element.get("tags", {})
        cafes.append({
            "name": tags.get("name", "Desconocido"),
            "official_name": tags.get("official_name", ""),
            "brand": tags.get("brand", ""),
            "street": tags.get("addr:street", ""),
            "housenumber": tags.get("addr:housenumber", ""),
            "city": tags.get("addr:city", ""),
            "postcode": tags.get("addr:postcode", ""),
            "operator": tags.get("operator", ""),
            "cuisine": tags.get("cuisine", ""),
            "internet_access": tags.get("internet_access", ""),
            "wheelchair": tags.get("wheelchair", ""),
            "toilets:wheelchair": tags.get("toilets:wheelchair", ""),
            "takeaway": tags.get("takeaway", ""),
            "opening_hours": tags.get("opening_hours", ""),
            "website": tags.get("website", ""),
            "lat": element["lat"],
            "lon": element["lon"]
        })
```

```python
print(cafes)
```

```python
len(cafes)
```

```python
#convetir a dataframe
df_cafe = pd.DataFrame(cafes)
print(df_cafe)
```

```python
gdf_starbucks = gpd.GeoDataFrame(df_cafe,
                                  geometry=gpd.points_from_xy(df_cafe.lon, df_cafe.lat),
                                  crs="EPSG:4326")
```

```python
import folium

# Crear un mapa centrado en Lima, Perú
m = folium.Map(location=[-12.0464, -77.0428], zoom_start=12)

# Mostrar el mapa si estás en un entorno compatible como Jupyter Notebook
m
```

```python
import folium
```

```python
from folium.plugins import MarkerCluster
```

```python
# Crear el grupo de clústeres
marker_cluster = MarkerCluster().add_to(m)
```

```python
#crear un mapa de Folium y añadir un clúster de marcadores
for idx, row in gdf_starbucks.iterrows():
    popup_text = f"""
    <b>{row['name']}</b><br>
    Dirección: {row['street']} {row['housenumber']}, {row['city']}<br>
    Horario: {row['opening_hours']}<br>
    Wifi: {row['internet_access']}<br>
    Takeaway: {row['takeaway']}<br>
    Accesibilidad: {row['wheelchair']}<br>
    <a href="{row['website']}" target="_blank">Sitio web</a>
    """
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(icon='coffee', prefix='fa')
    ).add_to(marker_cluster)
```
